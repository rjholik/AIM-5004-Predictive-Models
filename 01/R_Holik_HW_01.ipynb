Next, treat these samples as a data set for which you don't yet know  the true value of mu (pretend you don't know that mu=1) and must estimate it. #### Importing the necessary libraries

```python
import numpy as np
np.random.seed(2023)
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

np.set_printoptions(edgeitems=25, threshold=500, linewidth=500)
```

```python
# Error with normal distribution : p() ~ N(0, 1)
epsilon = np.random.normal(0, 1, 1000)
```

```python
# Samples (data set) of y with  = 1
mu = 1
y = mu + epsilon
```

```python
# Sample mean
muHat = np.mean(y)
print(f"^ = {muHat}")
```

$$
\hat{\mu} = 0.9929411990642036
$$

```python
# Range of candidate values of 
mu_candidates = np.arange(0, 2.01, .1)
print(f"mu_candidates = {mu_candidates}")
```

```python
# SEE for each candidate of 
SEE = []
for mu_candidate in mu_candidates:
    SEE.append((y - mu_candidate).T @ (y - mu_candidate))
```

```python
print(f"SEE = \n{SEE}")
```

```python
plt.plot(mu_candidates, SEE)
plt.xlabel(r"$\hat{\mu}$")
plt.ylabel(r"$SSE$")
plt.title("SSE vs mu_candidates")
plt.show()
```

From the above plot, it can be clearly deduced that the value $\hat{\mu} = 1$ minimizes SSE.

__<b>(b)</b> Notice that the households are ordered by the values of Consumers/Gardener (X). Divide the 20 households into three groups, placing the first seven households in the first group, the next six in the second group, and the last seven in the third group. Calculate the mean Y and mean X in each of the three groups. Transfer these means to the scatterplot and connect them with a simple nonparametric regression line. Does the regression line help you to interpret the relationship between the variables?__

```python
group1 = sahlinsData[:7]
group2 = sahlinsData[7:13]
group3 = sahlinsData[13:]
```

```python
group1
```

```python
group2
```

```python
group3
```

```python
meanGroupX = [group1["consumers"].mean(), group2["consumers"].mean(), group3["consumers"].mean()]
meanGroupY = [group1["acres"].mean(), group2["acres"].mean(), group3["acres"].mean()]
```

```python
# Linear regression model:

meanX = np.mean(meanGroupX)
meanY = np.mean(meanGroupY)
Cov_XY = np.sum((meanGroupX - meanX)*(meanGroupY - meanY))/(len(meanGroupX) - 1) 
Var_XX = np.sum((meanGroupX - meanX)**2)/(len(meanGroupX) - 1)

# Variance before model
TSS = np.sum((meanGroupY - meanY)**2)
Var_YY = TSS/(len(meanGroupY) - 1)

Beta = Cov_XY / Var_XX
Alpha = meanY - Beta * meanX
print(f"^y = {Alpha} + {Beta} x")


# estimate of y for x: ^y
y_hat = Beta * np.array(meanGroupX) + Alpha

# Variance after model
RSS = np.sum((meanGroupY - y_hat)**2)
print(f"RSS = {RSS}")

# Portion explained R, R  [0, 1], higher is better
R_squared = (TSS - RSS)/TSS
print(f"R = {R_squared}")
```

### $$\hat{y} = 1.3051572341367832 + 0.5561127894178417 x$$
### $$RSS = 0.05877208526048186$$
### $$R^2 = 0.5698469908162994$$

```python
plt.scatter(meanGroupX, meanGroupY, label = 'Means')
plt.plot(meanGroupX, Beta * np.array(meanGroupX) + Alpha, c = 'r', label = 'Regression Line')
plt.xlabel(colNames[0].title())
plt.ylabel(colNames[1].title())
plt.title("Acres/Gardener (Y) versus Consumers/Gardener (X)")
plt.legend()
plt.grid()
plt.show()
```

We can conclude from the above plot that the regression line demonstrates a somewhat positive linear relationship between the individual means of the given three groups. It is worth noting that $R^2$ (0.57) does not have a value close to one.
